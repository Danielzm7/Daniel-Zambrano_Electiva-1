```python
import pandas as pd
import numpy as np
```

```python
micron = pd.read_csv("https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/EMICRON/emicron_identificacion.csv")
```

```python
micron
```

```python
#1- Calcule la cantidad de micronegocios que hay en el departamento del Quindío encabezados por hombres y mujeres entre los 18 y los 28 años 
micron_quindio = micron[micron['COD_DEPTO'] == 63]
microneg_quindio = micron_quindio[(micron_quindio['P241'] >= 18) & (micron_quindio['P241'] <= 28)]['F_EXP'].sum()
print(f"Cantidad de micronegocios en el Quindío encabezados por personas entre 18 y 28 años: {microneg_quindio}")
int(microneg_quindio)
```

```python
#2 Cual es la ciudad en la que existe más micronegocios encabezados por hombres entre los 20 y 30 años 
micron_hombres = micron[(micron['P35'] == 1) & (micron['P241'] >= 20) & (micron['P241'] <= 30)]
micron_area_sum = micron.groupby('AREA')['F_EXP'].sum()
micron_area_sum_1 = micron_area_sum.sort_values(ascending=False)
micron_area_sum_1
```

```python
#3- En el departamento del Tolima cuantos micronegocios existen encabezados por mujeres que no tienen una persona que les ayude en el micronegocio
micron_tolima = micron[micron['COD_DEPTO'] == 73]
microneg_tolima = micron_tolima[(micron_tolima['P3031'] == 2)]['F_EXP'].sum()
print(f"Cantidad de micronegocios encabezados por mujeres en el Tolima sin ayudante: {microneg_tolima}")
int(microneg_tolima)
```

```python
#4- Cuantos micronegocios hay en la ciudad de Bogotá encabezados por mujeres que estén en el sector de la industria manufacturera 
micron_bogota = micron[micron['AREA'] == 11]
microneg_bogota = micron_bogota[(micron_bogota['P3031'] == 2)]['F_EXP'].sum()
print(f"Cantidad de micronegocios encabezados por mujeres en el Tolima sin ayudante: {microneg_tolima}")
int(microneg_tolima)
```

```python
# Bogotá tiene AREA == 11, mujeres P35 == 2, industria manufacturera GRUPOS4 == 2
micron_bogota_mujeres_manuf = micron[(micron['AREA'] == 11) & (micron['P35'] == 2) & (micron['GRUPOS4'] == 2)]
microneg_bogota_mujeres_manuf = micron_bogota_mujeres_manuf['F_EXP'].sum()
print(f"Cantidad de micronegocios en Bogotá encabezados por mujeres en industria manufacturera: {microneg_bogota_mujeres_manuf}")
int(microneg_bogota_mujeres_manuf)
```
