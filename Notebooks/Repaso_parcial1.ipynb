```python
import pandas as pd
import numpy as np
```

```python
emicron = pd.read_csv("https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/EMICRON/emicron_identificacion.csv")
```

```python
emicron
```

```python
#1- Calcule la cantidad de micronegocios que hay en el departamento del QuindÃ­o encabezados por hombres y mujeres entre los 18 y los 28 aÃ±os 
emicron_quindio = emicron[emicron['COD_DEPTO'] == 63]
emicroneg_quindio = emicron_quindio[(emicron_quindio['P241'] >= 18) & (emicron_quindio['P241'] <= 28)]['F_EXP'].sum()
print(f"Cantidad de micronegocios en el QuindÃ­o encabezados por personas entre 18 y 28 aÃ±os: {emicroneg_quindio}")
int(emicroneg_quindio)
```

```python
#2 Cual es la ciudad en la que existe mÃ¡s micronegocios encabezados por hombres entre los 20 y 30 aÃ±os 
# Filtrar micronegocios encabezados por hombres entre 20 y 30 aÃ±os
micron_hombres = emicron.query("P35 == 1 and 20 <= P241 <= 30")

# Agrupar por ciudad (o Ã¡rea) y sumar el factor de expansiÃ³n
micron_por_ciudad = micron_hombres.groupby('AREA')['F_EXP'].sum().sort_values(ascending=False)

# Mostrar la ciudad con mayor cantidad estimada de micronegocios
ciudad_top = micron_por_ciudad.idxmax()
cantidad_top = int(micron_por_ciudad.max())

print(f"La ciudad con mÃ¡s micronegocios encabezados con propietarios hombres entre 20 y 30 aÃ±os es: {ciudad_top} ({cantidad_top} micronegocios)")
```

```python
#3- En el departamento del Tolima cuantos micronegocios existen encabezados por mujeres que no tienen una persona que les ayude en el micronegocio
# Filtrar micronegocios en Tolima encabezados por mujeres sin ayudante
microneg_tolima = emicron.query("COD_DEPTO == 73 and P35 == 2 and P3031 == 2")['F_EXP'].sum()

# Mostrar el resultado como nÃºmero entero
print(f"Total de mujeres que tienen su micronegocio en Tolima sin ayudante: {int(microneg_tolima)}")
```

```python
#4- Cuantos micronegocios hay en la ciudad de BogotÃ¡ encabezados por mujeres que estÃ©n en el sector de la industria manufacturera 
emicron_bogota_mujeres_manuf = emicron[(emicron['AREA'] == 11) & (emicron['P35'] == 2) & (emicron['GRUPOS4'] == 2)]
emicroneg_bogota_mujeres_manuf = emicron_bogota_mujeres_manuf['F_EXP'].sum()
print(f"Cantidad de micronegocios en BogotÃ¡ encabezados por mujeres en industria manufacturera: {emicroneg_bogota_mujeres_manuf}")
int(emicroneg_bogota_mujeres_manuf)
```

```python
#5- Â¿QuÃ© gÃ©nero de propietario de micronegocio paga mÃ¡s prestaciones sociales?

# Filtrar micronegocios que sÃ­ pagan prestaciones sociales (P3032_1 == 1)
pagan_prestaciones = emicron[emicron['P3032_1'] == 1]

# Agrupar por gÃ©nero del propietario (P35) y sumar el factor de expansiÃ³n (F_EXP)
prestaciones_por_genero = pagan_prestaciones.groupby('P35')['F_EXP'].sum()

# Asignar etiquetas a los gÃ©neros (1: Hombre, 2: Mujer)
prestaciones_por_genero.index = prestaciones_por_genero.index.map({1: 'Hombres', 2: 'Mujeres'})

# Encontrar el gÃ©nero que mÃ¡s paga
genero_mas_paga = prestaciones_por_genero.idxmax()
cantidad_maxima = int(prestaciones_por_genero.max())

print("Cantidad de micronegocios que pagan prestaciones sociales, por gÃ©nero del propietario:")
print(prestaciones_por_genero.apply(int))
print("\n" + "="*50)
print(f"El gÃ©nero de propietario que mÃ¡s paga prestaciones sociales es: {genero_mas_paga}")
```
